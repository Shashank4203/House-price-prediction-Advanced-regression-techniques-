```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
```

```python
test = pd.read_csv('test.csv')
```

```python
dataset = pd.concat([train,test],axis=0)
```

```python
dataset.shape
```

```python
dataset
```

```python
categorical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtypes == 'O']
```

```python
for feature in categorical_with_nan:
    print('{} :       {}% of missing value'.format(feature,np.round(dataset[feature].isnull().mean()*100,2)))
```

```python
def replace_cat_nan(dataset,feature_with_nan):
    data = dataset.copy()
    data[feature_with_nan] = data[feature_with_nan].fillna('Missing')
    return data
dataset=replace_cat_nan(dataset,categorical_with_nan)

dataset[categorical_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtypes != 'O']
```

```python
numerical_with_nan.remove('SalePrice')
```

```python
for features in numerical_with_nan:
    print('{}---------->{}% of missing value.'.format(features,np.round(dataset[features].isnull().mean()*100,3)) )
```

```python
for feature in numerical_with_nan:
    median_value = dataset[feature].median()
    dataset[feature+'_nan'] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
```

```python
dataset[numerical_with_nan].isnull().sum()
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature] = dataset['YrSold']-dataset[feature]
```
